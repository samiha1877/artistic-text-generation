```python
!pip install peft evaluate rouge_score seaborn --quiet

import os
import random
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
from transformers import (
    AutoTokenizer, 
    AutoModelForCausalLM, 
    AdamW, 
    get_linear_schedule_with_warmup
)
from datasets import Dataset as HFDataset
import evaluate
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.auto import tqdm
import gc
import warnings
from peft import get_peft_model, LoraConfig, TaskType

warnings.filterwarnings('ignore')

os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'

def set_seed(seed):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

set_seed(42)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

torch.cuda.empty_cache()
gc.collect()

df = pd.read_csv('/kaggle/input/gutenberg-poetry-dataset/Gutenberg-Poetry.csv')

poems = df.groupby('gid')['s'].apply(lambda x: ' '.join(x)).reset_index()
poems.rename(columns={'s': 'text'}, inplace=True)

print("Dataset Overview:")
print(poems.head())

poems['length'] = poems['text'].apply(lambda x: len(x.split()))
plt.figure(figsize=(10,6))
sns.histplot(poems['length'], bins=50, kde=True, color='skyblue')
plt.title('Distribution of Poem Lengths')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.show()

poems['length_category'] = pd.cut(poems['length'], bins=5, labels=False)
plt.figure(figsize=(8,5))
sns.countplot(x='length_category', data=poems, palette='viridis')
plt.title('Poem Length Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()

hf_dataset = HFDataset.from_pandas(poems[['text']])

train_size = 0.9
train_dataset, val_dataset = hf_dataset.train_test_split(train_size=train_size, seed=42).values()

print(f"Training samples: {len(train_dataset)}")
print(f"Validation samples: {len(val_dataset)}")

tokenizer = AutoTokenizer.from_pretrained('/kaggle/input/llama-3.2/transformers/1b-instruct/1')

if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
    print("Pad token was not set. Using eos_token as pad_token.")
else:
    print(f"Pad token is set to: {tokenizer.pad_token}")

def tokenize_function(examples):
    return tokenizer(
        examples['text'], 
        padding='max_length', 
        truncation=True, 
        max_length=256
    )

tokenized_train = train_dataset.map(
    tokenize_function, 
    batched=True, 
    remove_columns=['text']
)
tokenized_val = val_dataset.map(
    tokenize_function, 
    batched=True, 
    remove_columns=['text']
)

tokenized_train.set_format('torch')
tokenized_val.set_format('torch')

batch_size = 4
train_loader = DataLoader(
    tokenized_train, 
    batch_size=batch_size, 
    shuffle=True, 
    num_workers=2
)
val_loader = DataLoader(
    tokenized_val, 
    batch_size=batch_size, 
    shuffle=False, 
    num_workers=2
)

model = AutoModelForCausalLM.from_pretrained(
    '/kaggle/input/llama-3.2/transformers/1b-instruct/1',
    torch_dtype=torch.float16,             
    low_cpu_mem_usage=True                 
)

if tokenizer.pad_token_id != model.config.pad_token_id:
    model.config.pad_token_id = tokenizer.pad_token_id
    model.resize_token_embeddings(len(tokenizer))
    print("Resized token embeddings to accommodate the new pad_token.")

lora_config = LoraConfig(
    task_type=TaskType.CAUSAL_LM,
    inference_mode=False,
    r=8,
    lora_alpha=32,
    lora_dropout=0.1,
)

model = get_peft_model(model, lora_config)

model.to(device)

print("Trainable parameters after applying LoRA:")
model.print_trainable_parameters()

if torch.cuda.device_count() > 1:
    print(f"Using {torch.cuda.device_count()} GPUs")
    model = nn.DataParallel(model)

optimizer = AdamW(filter(lambda p: p.requires_grad, model.parameters()), lr=5e-5)
num_epochs = 3
total_steps = num_epochs * len(train_loader)
scheduler = get_linear_schedule_with_warmup(
    optimizer, 
    num_warmup_steps=int(0.1 * total_steps), 
    num_training_steps=total_steps
)

train_losses = []
val_losses = []

bleu_metric = evaluate.load('bleu')
rouge_metric = evaluate.load('rouge')

scaler = torch.cuda.amp.GradScaler()

for epoch in range(num_epochs):
    model.train()
    epoch_train_loss = 0
    progress_bar = tqdm(train_loader, desc=f"Training Epoch {epoch+1}/{num_epochs}")
    
    for batch in progress_bar:
        inputs = batch['input_ids'].to(device, dtype=torch.long)
        attention_mask = batch['attention_mask'].to(device, dtype=torch.long)
        labels = batch['input_ids'].to(device, dtype=torch.long)
        
        optimizer.zero_grad()
        
        with torch.cuda.amp.autocast():
            outputs = model(input_ids=inputs, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss.mean()
        
        scaler.scale(loss).backward()
        
        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        
        scaler.step(optimizer)
        scaler.update()
        scheduler.step()
        
        epoch_train_loss += loss.item()
        progress_bar.set_postfix({'loss': loss.item()})
        
        torch.cuda.empty_cache()
    
    avg_train_loss = epoch_train_loss / len(train_loader)
    train_losses.append(avg_train_loss)
    
    model.eval()
    epoch_val_loss = 0
    with torch.no_grad():
        for batch in tqdm(val_loader, desc=f"Validation Epoch {epoch+1}/{num_epochs}"):
            inputs = batch['input_ids'].to(device, dtype=torch.long)
            attention_mask = batch['attention_mask'].to(device, dtype=torch.long)
            labels = batch['input_ids'].to(device, dtype=torch.long)
            
            with torch.cuda.amp.autocast():
                outputs = model(input_ids=inputs, attention_mask=attention_mask, labels=labels)
                loss = outputs.loss.mean()
            epoch_val_loss += loss.item()
            
            torch.cuda.empty_cache()
    
    avg_val_loss = epoch_val_loss / len(val_loader)
    val_losses.append(avg_val_loss)
    
    print(f"Epoch {epoch+1} | Train Loss: {avg_train_loss:.4f} | Val Loss: {avg_val_loss:.4f}")
    
    checkpoint_dir = f'./model_epoch_{epoch+1}_lora'
    if isinstance(model, nn.DataParallel):
        model.module.save_pretrained(checkpoint_dir)
    else:
        model.save_pretrained(checkpoint_dir)
    tokenizer.save_pretrained(checkpoint_dir)
    
    print(f"Model checkpoint saved to {checkpoint_dir}")
    
    del inputs, attention_mask, labels, outputs
    torch.cuda.empty_cache()
    gc.collect()

plt.figure(figsize=(10,6))
plt.plot(range(1, num_epochs+1), train_losses, marker='o', label='Train Loss')
plt.plot(range(1, num_epochs+1), val_losses, marker='o', label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.grid(True)
plt.show()

model.eval()
sample_size = 100
references = []
predictions = []

for i, batch in enumerate(val_loader):
    if i * batch_size >= sample_size:
        break
    inputs = batch['input_ids'].to(device, dtype=torch.long)
    attention_mask = batch['attention_mask'].to(device, dtype=torch.long)
    
    with torch.no_grad():
        if isinstance(model, nn.DataParallel):
            generated_ids = model.module.generate(
                input_ids=inputs, 
                attention_mask=attention_mask, 
                max_new_tokens=100, 
                num_beams=5,
                no_repeat_ngram_size=2,
                early_stopping=True
            )
        else:
            generated_ids = model.generate(
                input_ids=inputs, 
                attention_mask=attention_mask, 
                max_new_tokens=100, 
                num_beams=5,
                no_repeat_ngram_size=2,
                early_stopping=True
            )
    
    preds = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)
    refs = tokenizer.batch_decode(inputs, skip_special_tokens=True)
    
    predictions.extend(preds)
    references.extend([[ref] for ref in refs])
    
    torch.cuda.empty_cache()

bleu_result = bleu_metric.compute(predictions=predictions, references=references)
print(f"BLEU Score: {bleu_result['bleu']:.4f}")
```

```python
!pip install peft evaluate rouge_score seaborn --quiet

import os
import random
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
from transformers import (
    AutoTokenizer, 
    AutoModelForCausalLM, 
    AdamW, 
    get_linear_schedule_with_warmup
)
from datasets import Dataset as HFDataset
import evaluate
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.auto import tqdm
import gc
import warnings
from peft import get_peft_model, LoraConfig, TaskType

warnings.filterwarnings('ignore')

os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'

def set_seed(seed):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

set_seed(42)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

torch.cuda.empty_cache()
gc.collect()

df = pd.read_csv('/kaggle/input/gutenberg-poetry-dataset/Gutenberg-Poetry.csv')

poems = df.groupby('gid')['s'].apply(lambda x: ' '.join(x)).reset_index()
poems.rename(columns={'s': 'text'}, inplace=True)

print("Dataset Overview:")
print(poems.head())

poems['length'] = poems['text'].apply(lambda x: len(x.split()))
plt.figure(figsize=(10,6))
sns.histplot(poems['length'], bins=50, kde=True, color='skyblue')
plt.title('Distribution of Poem Lengths')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.show()

poems['length_category'] = pd.cut(poems['length'], bins=5, labels=False)
plt.figure(figsize=(8,5))
sns.countplot(x='length_category', data=poems, palette='viridis')
plt.title('Poem Length Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()

hf_dataset = HFDataset.from_pandas(poems[['text']])

train_size = 0.9
train_dataset, val_dataset = hf_dataset.train_test_split(train_size=train_size, seed=42).values()

print(f"Training samples: {len(train_dataset)}")
print(f"Validation samples: {len(val_dataset)}")

tokenizer = AutoTokenizer.from_pretrained('/kaggle/input/llama-3.2/transformers/1b-instruct/1')

if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
    print("Pad token was not set. Using eos_token as pad_token.")
else:
    print(f"Pad token is set to: {tokenizer.pad_token}")

def tokenize_function(examples):
    return tokenizer(
        examples['text'], 
        padding='max_length', 
        truncation=True, 
        max_length=256
    )

tokenized_train = train_dataset.map(
    tokenize_function, 
    batched=True, 
    remove_columns=['text']
)
tokenized_val = val_dataset.map(
    tokenize_function, 
    batched=True, 
    remove_columns=['text']
)

tokenized_train.set_format('torch')
tokenized_val.set_format('torch')

batch_size = 8
train_loader = DataLoader(
    tokenized_train, 
    batch_size=batch_size, 
    shuffle=True, 
    num_workers=2
)
val_loader = DataLoader(
    tokenized_val, 
    batch_size=batch_size, 
    shuffle=False, 
    num_workers=2
)

model = AutoModelForCausalLM.from_pretrained(
    '/kaggle/input/llama-3.2/transformers/1b-instruct/1',
    torch_dtype=torch.float16,
    low_cpu_mem_usage=True
)

if tokenizer.pad_token_id != model.config.pad_token_id:
    model.config.pad_token_id = tokenizer.pad_token_id
    model.resize_token_embeddings(len(tokenizer))
    print("Resized token embeddings to accommodate the new pad_token.")

lora_config = LoraConfig(
    task_type=TaskType.CAUSAL_LM,
    inference_mode=False,
    r=16,
    lora_alpha=64,
    lora_dropout=0.1,
)

model = get_peft_model(model, lora_config)

model.to(device)

print("Trainable parameters after applying LoRA:")
model.print_trainable_parameters()

if torch.cuda.device_count() > 1:
    print(f"Using {torch.cuda.device_count()} GPUs")
    model = nn.DataParallel(model)

optimizer = AdamW(filter(lambda p: p.requires_grad, model.parameters()), lr=3e-5)
num_epochs = 10
total_steps = num_epochs * len(train_loader)
scheduler = get_linear_schedule_with_warmup(
    optimizer, 
    num_warmup_steps=int(0.1 * total_steps), 
    num_training_steps=total_steps
)

train_losses = []
val_losses = []

bleu_metric = evaluate.load('bleu')
rouge_metric = evaluate.load('rouge')

scaler = torch.cuda.amp.GradScaler()

for epoch in range(num_epochs):
    model.train()
    epoch_train_loss = 0
    progress_bar = tqdm(train_loader, desc=f"Training Epoch {epoch+1}/{num_epochs}")
    
    for batch in progress_bar:
        inputs = batch['input_ids'].to(device, dtype=torch.long)
        attention_mask = batch['attention_mask'].to(device, dtype=torch.long)
        labels = batch['input_ids'].to(device, dtype=torch.long)
        
        optimizer.zero_grad()
        
        with torch.cuda.amp.autocast():
            outputs = model(input_ids=inputs, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss.mean()
        
        scaler.scale(loss).backward()
        
        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        
        scaler.step(optimizer)
        scaler.update()
        scheduler.step()
        
        epoch_train_loss += loss.item()
        progress_bar.set_postfix({'loss': loss.item()})
        
        torch.cuda.empty_cache()
    
    avg_train_loss = epoch_train_loss / len(train_loader)
    train_losses.append(avg_train_loss)
    
    model.eval()
    epoch_val_loss = 0
    with torch.no_grad():
        for batch in tqdm(val_loader, desc=f"Validation Epoch {epoch+1}/{num_epochs}"):
            inputs = batch['input_ids'].to(device, dtype=torch.long)
            attention_mask = batch['attention_mask'].to(device, dtype=torch.long)
            labels = batch['input_ids'].to(device, dtype=torch.long)
            
            with torch.cuda.amp.autocast():
                outputs = model(input_ids=inputs, attention_mask=attention_mask, labels=labels)
                loss = outputs.loss.mean()
            epoch_val_loss += loss.item()
            
            torch.cuda.empty_cache()
    
    avg_val_loss = epoch_val_loss / len(val_loader)
    val_losses.append(avg_val_loss)
    
    print(f"Epoch {epoch+1} | Train Loss: {avg_train_loss:.4f} | Val Loss: {avg_val_loss:.4f}")
    
    checkpoint_dir = f'./model_epoch_{epoch+1}_lora'
    if isinstance(model, nn.DataParallel):
        model.module.save_pretrained(checkpoint_dir)
    else:
        model.save_pretrained(checkpoint_dir)
    tokenizer.save_pretrained(checkpoint_dir)
    
    print(f"Model checkpoint saved to {checkpoint_dir}")
    
    del inputs, attention_mask, labels, outputs
    torch.cuda.empty_cache()
    gc.collect()

plt.figure(figsize=(10,6))
plt.plot(range(1, num_epochs+1), train_losses, marker='o', label='Train Loss')
plt.plot(range(1, num_epochs+1), val_losses, marker='o', label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.grid(True)
plt.show()

model.eval()
sample_size = 200
references = []
predictions = []

for i, batch in enumerate(val_loader):
    if i * batch_size >= sample_size:
        break
    inputs = batch['input_ids'].to(device, dtype=torch.long)
    attention_mask = batch['attention_mask'].to(device, dtype=torch.long)
    
    with torch.no_grad():
        if isinstance(model, nn.DataParallel):
            generated_ids = model.module.generate(
                input_ids=inputs, 
                attention_mask=attention_mask, 
                max_new_tokens=150, 
                num_beams=5,
                no_repeat_ngram_size=2,
                early_stopping=True
            )
        else:
            generated_ids = model.generate(
                input_ids=inputs, 
                attention_mask=attention_mask, 
                max_new_tokens=150, 
                num_beams=5,
                no_repeat_ngram_size=2,
                early_stopping=True
            )
    
    preds = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)
    refs = tokenizer.batch_decode(inputs, skip_special_tokens=True)
    
    predictions.extend(preds)
    references.extend([[ref] for ref in refs])
    
    torch.cuda.empty_cache()

bleu_result = bleu_metric.compute(predictions=predictions, references=references)
print(f"BLEU Score: {bleu_result['bleu']:.4f}")

rouge_result = rouge_metric.compute(predictions=predictions, references=references, use_stemmer=True)
print(f"ROUGE-1: {rouge_result['rouge1']:.4f}")
print(f"ROUGE-2: {rouge_result['rouge2']:.4f}")
print(f"ROUGE-L: {rouge_result['rougeL']:.4f}")

metrics = {
    'BLEU': bleu_result['bleu'],
    'ROUGE-1': rouge_result['rouge1'],
    'ROUGE-2': rouge_result['rouge2'],
    'ROUGE-L': rouge_result['rougeL']
}

plt.figure(figsize=(8,6))
sns.barplot(x=list(metrics.keys()), y=list(metrics.values()), palette='magma')
plt.ylim(0,1)
plt.title('Evaluation Metrics')
plt.ylabel('Score')
for index, value in enumerate(metrics.values()):
    plt.text(index, value + 0.02, f"{value:.2f}", ha='center')
plt.show()

def generate_poem(prompt, max_new_tokens=150):
    model.eval()
    inputs = tokenizer(prompt, return_tensors='pt').to(device)
    with torch.no_grad():
        if isinstance(model, nn.DataParallel):
            generated_ids = model.module.generate(
                **inputs, 
                max_new_tokens=max_new_tokens, 
                num_return_sequences=1, 
                no_repeat_ngram_size=2, 
                early_stopping=True,
                temperature=0.7,
                top_p=0.9
            )
        else:
            generated_ids = model.generate(
                **inputs, 
                max_new_tokens=max_new_tokens, 
                num_return_sequences=1, 
                no_repeat_ngram_size=2, 
                early_stopping=True,
                temperature=0.7,
                top_p=0.9
            )
    poem = tokenizer.decode(generated_ids[0], skip_special_tokens=True)
    return poem

sample_prompts = [
    "The silent night",
    "Upon the morning dew",
    "Whispers of the wind",
    "Dancing shadows",
    "Eternal dreams"
]

for prompt in sample_prompts:
    print(f"Prompt: {prompt}")
    print(f"Generated Poem:\n{generate_poem(prompt)}\n{'-'*50}")

plt.figure(figsize=(10,6))
plt.plot(range(1, num_epochs+1), train_losses, marker='o', label='Train Loss')
plt.plot(range(1, num_epochs+1), val_losses, marker='o', label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Overfitting Check')
plt.legend()
plt.grid(True)
plt.show()

# ================== NEW CODE STARTS HERE ==================

print("Merging LoRA adapters into the base model for deployment...")
if isinstance(model, nn.DataParallel):
    model = model.module
model = model.merge_and_unload()
print("LoRA adapters merged successfully.")

finetuned_model_dir = './finetuned_model'
print(f"Saving the fine-tuned model to {finetuned_model_dir}...")
model.save_pretrained(finetuned_model_dir)
tokenizer.save_pretrained(finetuned_model_dir)
print("Fine-tuned model and tokenizer saved successfully.")

del model
torch.cuda.empty_cache()
gc.collect()

print("Loading the fine-tuned model for verification...")
from transformers import AutoModelForCausalLM, AutoTokenizer

loaded_model = AutoModelForCausalLM.from_pretrained(finetuned_model_dir)
loaded_tokenizer = AutoTokenizer.from_pretrained(finetuned_model_dir)
loaded_model.to(device)
print("Fine-tuned model loaded successfully.")

def generate_poem_deployed(prompt, max_new_tokens=150):
    loaded_model.eval()
    inputs = loaded_tokenizer(prompt, return_tensors='pt').to(device)
    with torch.no_grad():
        generated_ids = loaded_model.generate(
            **inputs, 
            max_new_tokens=max_new_tokens,
            num_return_sequences=1, 
            no_repeat_ngram_size=2, 
            early_stopping=True,
            temperature=0.7,
            top_p=0.9
        )
    poem = loaded_tokenizer.decode(generated_ids[0], skip_special_tokens=True)
    return poem

sample_prompts = [
    "The silent night",
    "Upon the morning dew",
    "Whispers of the wind",
    "Dancing shadows",
    "Eternal dreams"
]

for prompt in sample_prompts:
    print(f"Prompt: {prompt}")
    print(f"Generated Poem:\n{generate_poem_deployed(prompt)}\n{'-'*50}")

del loaded_model, loaded_tokenizer
torch.cuda.empty_cache()
gc.collect()

print("Fine-tuning complete with tuned hyperparameters. The complete fine-tuned model is saved and ready for deployment.")


```
